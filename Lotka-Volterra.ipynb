```python
from numpy import *
import pylab as p
# Definition of parameters
a = 1.
b = 0.5
c = 0.5
d = 2.
def dX_dt(X, t=0):
    """ Return the growth rate of fox and rabbit populations. """
    return array([ a*X[0] -   b*X[0]*X[1] ,
                  -d*X[1] +   c*X[0]*X[1] ])
```

```python
from scipy import integrate
t1 = linspace(0, 30,  30)  # time with interval of 1  
t2 = linspace(0, 30,  100)  # time with interval of 0.3
X0 = array([2, 2])                     # initials conditions: 2000 rabbits and 2000 foxes
X1 = integrate.odeint(dX_dt, X0, t1)
X2 = integrate.odeint(dX_dt, X0, t2)
```

```python
rabbits1, foxes1 = X1.T
f1 = p.figure()
p.plot(t1, rabbits1, 'b-', label='Rabbits')
p.plot(t1, foxes1  , 'r-', label='Foxes')
p.grid()
p.legend(loc='best')
p.xlabel('time')
p.ylabel('population')
p.title('Evolution of fox and rabbit populations')
f1.savefig('rabbits_and_foxes_1.png')
```

```python
rabbits2, foxes2 = X2.T
f2 = p.figure
p.plot(t2, rabbits2, 'b-', label='Rabbits')
p.plot(t2, foxes2  , 'r-', label='Foxes')
p.grid()
p.legend(loc='best')
p.xlabel('time')
p.ylabel('population')
p.title('Evolution of fox and rabbit populations')
```
